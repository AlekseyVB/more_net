### код для преобразования меток датасета на один лист

```python
def yolo_to_mot(yolo_data, file_name):
    # Разбиваем строку с данными YOLO на составляющие
    data = yolo_data.split(' ')
    if len(data) == 5:
        class_id, x, y, w, h = data
    else:
        class_id = data[0]
        x = 0.5
        y = 0.5
        w = 0.0
        h = 0.0

    # Вычисляем fbox (x1, y1, x2, y2)
    x1 = int(float(x) * 1280 - float(w) * 1280 / 2)
    y1 = int(float(y) * 1280 - float(h) * 1280 / 2)
    x2 = int(float(x) * 1280 + float(w) * 1280 / 2)
    y2 = int(float(y) * 1280 + float(h) * 1280 / 2)

    # Вычисляем hbox (x1, y1, w, h), vbox (x1, y1, w, h), head_attr (ignore=0 occ=0 unsure=0). 
    # Для удобства ignore=0 occ=0 unsure=0 помещены в extra.box_id. 
    hbox = [int((x2 + x1)/2), int((y2 + y1)/2), int((x2 - x1)), int((y2 - 1))]
    vbox = [int((x2 + x1)/2), int((y2 + y1)/2), int((x2 - x1)), int((y2 - 1))]

        
    # Формируем MOT-данные. ID указывать необязательно.  
    mot_data = {'ID': file_name , 'gtboxes': [{'fbox': [int(x1), int(y1), int(x2),\
            int(y2)],'tag': class_id, 'hbox': hbox,'extra': {'box_id': 0,'occ': 0},\
            'vbox': vbox,'head_attr': {'ignore': 0,'occ': 0,'unsure': 0}}]}

    return mot_data
```

```python
import os

wayTrainLabels = '/content/FairMOT/dataset/train/labels'
wayValLabels = '/content/FairMOT/dataset/val/labels'

# путь до директории с txt файлами
path = wayValLabels

# имя выходного файла
output_file = '/content/FairMOT/dataset/annotation_Val.odgt'


# открываем файл для записи результатов работы цикла
with open(output_file, 'w') as f:
    # перебираем txt файлы из директории path
    for file_yolo in os.listdir(path):
        # открываем txt файл, считываем содержимое, закрываем txt файл        
        with open(os.path.join(path, file_yolo), 'r') as f2:
            # перебор строк txt-файлов            
            for line in f2:
                # применяем yolo_to_mot() 
                print('yolo:',line)                                                                                                       
                line = yolo_to_mot(line, file_yolo.split('.')[0])
                print('MOT:',line)
                # запись строк (с yolo_to_mot())                                                        
                f.write('\n' + str(line))
```
