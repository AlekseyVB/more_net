```python
import os
import pandas as pd
train_dir = '/content/FairMOT/dataset/images/train/'
val_dir = '/content/FairMOT/dataset/images/val/'

training_files = os.listdir('/content/images/images/train')
val_files = os.listdir('/content/images/images/val')

training_df = pd.DataFrame(training_files,columns=['images'])
training_df.images = training_df.images.map(lambda x: train_dir + x)
val_df = pd.DataFrame(val_files,columns=['images'])
val_df.images = val_df.images.map(lambda x: val_dir + x)

%pwd
```

```python
#Generate files containing image paths and save to src/data
train_file = 'src/data/crowdhuman.train'
test_file = 'src/data/crowdhuman.val'
with open(train_file, 'w') as f:
    f.write("\n".join(training_df.images.values))
with open(test_file, 'w') as f:
    f.write("\n".join(val_df.images.values))
```

```python
#Create a json file for the dataset in src/lib/cfg/

import json
train_file = 'src/data/crowdhuman.train'
test_file  = 'src/data/crowdhuman.val'
cfg = dict(
    root='/content/FairMOT/dataset',
    train=dict(CrowdHuman_train01=train_file,
               CrowedHuman_test=test_file),
    test=dict(CrowedHuman_test=test_file),
    test_emb=dict(CrowedHuman_test=test_file),
)
with open('src/lib/cfg/crowdhuman.json', 'w') as f:
    json.dump(cfg, f, indent=4)
```

- Tot: 0:02:39: это общее время работы обучения в часах, минутах и секундах.
- ETA: 0:02:08: это предполагаемое время до завершения обучения. Значение определяется на основе общего времени обучения, на которое нужно потратить, и количества эпох, оставшихся для обучения.
- loss: это общая потеря (или ошибка) при обучении за текущую эпоху. Более низкое значение loss указывает на более эффективное обучение.
- hm_loss: это потеря при обучении модели обнаружения человека (HeatMap Loss).
- wh_loss: это потеря при обучении модели точного предсказания координат (Width Height Loss).
- off_loss: это потеря смещения при обучении модели точного предсказания смещения объектов (Offset Loss).
- id_loss: это потеря при обучении модели для идентификации объектов (ID Loss).
- Data 0.011s(0.217s): это время загрузки датасета на данной эпохе и среднее время загрузки данных для каждого батча соответственно.
- Net 1.208s: это время, которое модель затратила на обработку батча данных на данной эпохе.
